```python
import os
import nltk
import csv
nltk.download('punkt')
```

```python
# Step 1: Read in the corpus of files
corpus_dir = "E:/Professional Masters in Data Science/SEM 2/Machine Learning/Project/Poets/EmilyDickinson"
files = os.listdir(corpus_dir)
```

```python
stop_words = set(nltk.corpus.stopwords.words("english"))
```

```python
stop_words

```

```python
# Step 2: Tokenize the corpus
tokens = []
# Step 4: Convert the corpus into transaction data format
transactions = []
for file in files:
    with open(os.path.join(corpus_dir, file), "r") as f:
        text = f.read()
        file_tokens = [token.lower() for token in nltk.word_tokenize(text) if token.lower() not in stop_words and token.isalpha()]
        transactions.append(file_tokens)
```

```python
with open("transactions.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(transactions)
```
