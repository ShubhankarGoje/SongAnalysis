```python

# import necessary libraries
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd   # for dataframes
import os
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
url="https://www.poemhunter.com/poem/faith-is-a-fine-invention/"
title=""
#URL 1 to scrape web
```

```python
directory = "EmilyDickinson"
parent_dir = "E:/Professional Masters in Data Science/SEM 2/Machine Learning/Project"
path = os.path.join(parent_dir, directory)
os.mkdir(path)
#Creating a new corpus
```

```python
#Looping through all the poems and scraping and cleaning the text simultaniously while writing them


for i in range(1, 1232):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for e in soup.find_all('div',class_='phBoxContinue txtc phDetailAnother'):
        for link in e.find_all('a',attrs={'href': re.compile("/poem")}):
            url="https://www.poemhunter.com"+str(link.get('href'))
            
    for name in soup.find_all('meta',property="og:article:tag"):
        title=str(name.get('content')).replace("&Mdash;"," ")
        title=title.replace("\"","")
        title=re.sub(r'[^a-zA-Z]', " ", title, flags=re.VERBOSE)
        print(title)
    f = open(path+'/'+title+'.txt', 'w')
    for e in soup.find_all('div',class_='phContent phcText'):
        for t in list(e.stripped_strings):
            f.write(t.replace("—"," ")+"\n")
f.close()       
```

```python
#Cleating a list of file names with complete file path for count vertorizer
FullNameList=[]
LablesList=[]
for name in os.listdir(path):
    FullNameList.append(path+"/"+name)
    LablesList.append(re.split(r'[123456789.]+', name)[0])
print(FullNameList)
print(LablesList)
scrapedEDPoemDF=pd.DataFrame()
```

```python
#Count Vetorizer for Scraped Data

scrapedEDPoemCountVectorizer=CountVectorizer(input = "filename", stop_words="english",
                                             encoding="iso-8859-1")
scrapedEDPoemData = scrapedEDPoemCountVectorizer.fit_transform(FullNameList)
scrapedEDPoemColNames=scrapedEDPoemCountVectorizer.get_feature_names()
scrapedEDPoemDF=pd.DataFrame(scrapedEDPoemData.toarray(), columns=scrapedEDPoemColNames)
print(scrapedEDPoemDF.head())
```

```python
#Removing unnessesory data 

scrapedEDPoemDF.drop(scrapedEDPoemDF.iloc[:, 0:866],axis = 1,inplace=True)
```

```python

#Clean Bag of Words
scrapedEDPoemDF.to_csv("scrapedEDPoemDF.csv")
```

```python
#Gathering top 200 songs from artists
artist= genius.search_artist('Taylor Swift',max_songs=200)

```

```python

#Creating directory with Artist name 
directory = "TaylorSwift"
parent_dir = "E:/Professional Masters in Data Science/SEM 2/Machine Learning/Project"
pathT = os.path.join(parent_dir, directory)
os.mkdir(pathT)
```

```python
#Saving Lyrics with Song title as file name and Lyrics as content
for song in artist.songs:
    title=re.sub(r'[^a-zA-Z]', " ", song.title, flags=re.VERBOSE)
    f = open(pathT+'/'+title+'.txt', 'w' , encoding="utf-8")
    lyrics=re.sub(r'[^a-zA-Z]', " ", song.lyrics, flags=re.VERBOSE)
    f.write(lyrics)
    f.close()
```

```python
#Cleating a list of file names with complete file path for count vertorizer

FullNameList=[]
LablesList=[]
for name in os.listdir(pathT):
    FullNameList.append(pathT+"/"+name)
    LablesList.append(re.split(r'[.]+', name)[0])
print(FullNameList)
print(LablesList)
scrapedTSDF=pd.DataFrame()
```

```python
#Conut Vectorizer for APi Data
scrapedTSCountVectorizer=CountVectorizer(input = "filename", encoding='utf-8')
scrapedTSData = scrapedTSCountVectorizer.fit_transform(FullNameList)
scrapedTSColNames=scrapedTSCountVectorizer.get_feature_names()
scrapedTSDF=pd.DataFrame(scrapedTSData.toarray(), columns=scrapedTSColNames)
print(scrapedTSDF.head())
```

```python
#Saving Data as CSV

scrapedTSDF.to_csv("scrapedTSDF.csv")
```
